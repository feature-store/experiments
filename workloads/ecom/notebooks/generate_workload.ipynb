```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
train_df = pd.read_parquet("0031/train.parquet")
valid_df = pd.read_parquet("0031/valid.parquet")
test_df = pd.read_parquet("0031/test.parquet")
```

```python
df = pd.concat([train_df, valid_df, test_df], ignore_index=True)
df
```

```python
df = df[["user_idx", "sess_pid_seq", "sess_etime_seq"]]
df
```

```python
min_ts = df["sess_etime_seq"].explode().min()
df["sess_etime_seq"] = df["sess_etime_seq"].apply(lambda x: x - min_ts)
df
```

```python
df["sess_etime_seq"] = df["sess_etime_seq"].apply(lambda x: x - int(x[0] * 0.999))
df
```

```python
df["sess_etime_seq"].explode().value_counts(sort=True)
```

```python
start = 0
end = 300

raw_data = df.explode(["sess_pid_seq", "sess_etime_seq"])
truncated = raw_data[(raw_data["sess_etime_seq"] >= start) & (raw_data["sess_etime_seq"] <= end)]
truncated = truncated.sort_values(by="sess_etime_seq", ascending=True)
truncated = truncated.reset_index(drop=True)

truncated["sess_etime_seq"] = truncated["sess_etime_seq"].apply(lambda x: x - start)
truncated
```

```python
# Create the event stream
event_stream = truncated[["user_idx", "sess_etime_seq", "sess_pid_seq"]]
event_stream.rename({"user_idx": "key_id", "sess_etime_seq": "ts", "sess_pid_seq": "value"}, axis=1, inplace=True)
event_stream = event_stream.rename_axis("event_id")
event_stream.to_csv("events.csv")
```

```python
# Create the query stream

# Remove the last interaction per user
df["sess_etime_seq"] = df["sess_etime_seq"].apply(lambda x: x[:-1])
df["sess_pid_seq"] = df["sess_pid_seq"].apply(lambda x: x[:-1])

raw_data = df.explode(["sess_pid_seq", "sess_etime_seq"])
truncated = raw_data[(raw_data["sess_etime_seq"] >= start) & (raw_data["sess_etime_seq"] <= end)]
truncated = truncated.sort_values(by="sess_etime_seq", ascending=True)
truncated = truncated.reset_index(drop=True)

truncated["sess_etime_seq"] = truncated["sess_etime_seq"].apply(lambda x: x - start)
truncated

query_stream = truncated[["user_idx", "sess_etime_seq"]]
query_stream.rename({"user_idx": "key_id", "sess_etime_seq": "ts"}, axis=1, inplace=True)
query_stream["ts"] = query_stream["ts"].apply(lambda x: x + 1)
query_stream = query_stream.rename_axis("query_id")
query_stream.to_csv("queries.csv")
```

```python
qps = truncated["sess_etime_seq"].value_counts(sort=False).to_numpy()
plt.plot(list(range(0, end - start + 1)), qps, color="red")
# plt.savefig("qps.png", dpi=1200)
```
